```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://fbref.com/en/squads/206d90db/2021-2022/matchlogs/all_comps/schedule/Barcelona-Scores-and-Fixtures-All-Competitions')

soup = BeautifulSoup(res.text, 'lxml')
# print(soup.find_all('table')[0].prettify())
```

```python
import pandas as pd

df = pd.read_html(soup.find_all('table')[0].prettify())[0]
```

```python
import re

exp = re.compile(r"^[a-z]{2}\s(.*)")

def remove_country(team):
    regex = exp.findall(team)
    if regex:
        return regex[0].strip()
    return team

df["Opponent"] = df["Opponent"].apply(remove_country)
df.head()
```

```python
def parse_team_stat(soup, team, stat_id):
    html_str = str(soup.find(id='team_stats'))
    df_team_stat = pd.read_html(str(soup.find(id='team_stats')), header=0)[0]

    stat_dict = {
    'club': team,
    }

    # get formation
    field = str(soup.find(id='field_wrap').prettify())
    formation = re.findall(fr'{team} \((.*)\)', field)[0]
    stat_dict['formation'] = formation

    possession =  df_team_stat[team].iloc[1]
    possession = float(possession.replace('%',''))
    stat_dict['possession'] = possession

    if len(df_team_stat[df_team_stat[team] == 'Passing Accuracy']):
        index = df_team_stat[df_team_stat[team] == 'Passing Accuracy'].index.values[0]
        passing_accuracy = df_team_stat[team].iloc[index + 1]

        successful_pass, total_pass_attempt = re.findall(r'(\d*) of (\d*)', f' {passing_accuracy}')[0]
        total_pass_attempt = int(total_pass_attempt)

        if successful_pass:
            successful_pass = int(successful_pass)
        else:
            successful_pass = 0

        if total_pass_attempt:
            passing_accuracy_percentage = round((successful_pass/total_pass_attempt) * 100, 2)
        else:
            passing_accuracy_percentage = 0

        stat_dict['passing_accuracy_percentage'] = passing_accuracy_percentage
        stat_dict['total_pass_attempt'] = int(total_pass_attempt)
        stat_dict['successful_pass'] = int(successful_pass)
    
    
    if len(df_team_stat[df_team_stat[team] == 'Shots on Target']):
        index = df_team_stat[df_team_stat[team] == 'Shots on Target'].index.values[0]
        shots_on_target = df_team_stat[team].iloc[index + 1]

        successful_shots_on_target, shots_on_target_attempt = re.findall(r'(\d*) of (\d*)', f' {shots_on_target}')[0]
        shots_on_target_attempt = int(shots_on_target_attempt)

        if successful_shots_on_target:
            successful_shots_on_target = int(successful_shots_on_target)
        else:
            successful_shots_on_target = 0

        if shots_on_target_attempt:
            shots_on_target_percentage = round((successful_shots_on_target/shots_on_target_attempt) * 100, 2)
        else:
            shots_on_target_percentage = 0

        stat_dict['shots_on_target_percentage'] = shots_on_target_percentage
        stat_dict['shots_on_target_attempt'] = int(shots_on_target_attempt)
        stat_dict['successful_shots_on_target'] = int(successful_shots_on_target)

    if len(df_team_stat[df_team_stat[team] == 'Saves']):
        index = df_team_stat[df_team_stat[team] == 'Saves'].index.values[0]
        saves = df_team_stat[team].iloc[index + 1]

        successful_saves, saves_attempt = re.findall(r'(\d*) of (\d*)', f' {saves}')[0]
        saves_attempt = int(saves_attempt)

        if successful_saves:
            successful_saves = int(successful_saves)
        else:
            successful_saves = 0

        if saves_attempt:
            saves_percentage = round((successful_saves/saves_attempt) * 100, 2)
        else:
            saves_percentage = 0

        stat_dict['saves_percentage'] = saves_percentage
        stat_dict['saves_attempt'] = int(saves_attempt)
        stat_dict['successful_saves'] = int(successful_saves)        

    yellow_card = len(soup.select(f'#team_stats > table > tr:last-child > td:nth-child({stat_id}) > div > div > div > .yellow_card'))
    yellow_red_card = len(soup.select(f'#team_stats > table > tr:last-child > td:nth-child({stat_id}) > div > div > div > .yellow_red_card'))
    red_card = len(soup.select(f'#team_stats > table > tr:last-child > td:nth-child({stat_id}) > div > div > div > .red_card'))

    stat_dict['yellow_card'] = yellow_card
    stat_dict['yellow_red_card'] = yellow_red_card
    stat_dict['red_card'] = red_card

    for i in soup.select('#team_stats_extra > div'):
        for j in re.findall(r'(?m)\n?(\d+)?([a-z A-Z]+)?(\d+)\n', i.text):
            if stat_id == 1:
                index = 0
            else:
                index = 2
            stat_dict[j[1].lower()] = int(j[index])

    return stat_dict
    
```

```python
import time

list_stat = []
for opponent, home_away, match_report_link in zip(df['opponent'], df['venue'], df['match_report']):
    res = requests.get(match_report_link)
    soup_report = BeautifulSoup(res.text, 'lxml')
    
    if home_away == 'Home':
        list_stat.append(parse_team_stat(soup_report, 'Barcelona', 1))
        list_stat.append(parse_team_stat(soup_report, opponent, 2))
        print(f'Barcelona vs {opponent}')
    else:
        list_stat.append(parse_team_stat(soup_report, opponent, 1))
        list_stat.append(parse_team_stat(soup_report, 'Barcelona', 2))
        print(f'{opponent} vs Barcelona')
    
    time.sleep(5)
    # break
    
df_stat = pd.DataFrame.from_dict(list_stat)
```
